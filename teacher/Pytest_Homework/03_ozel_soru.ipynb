Bir oyun kafesindeki oyuncuları ve görevleri yöneten bir Python programı yazınız.

Başlangıçta aşağıdaki yapıları oluşturun:

1) Oyuncular dictionary'si:
- Her oyuncu için şu bilgiler bulunsun:
  - seviye (int)
  - puan (int)
  - envanter (list)
  - aktif (bool)

2) Görevler listesi:
- Her görev bir dictionary olsun ve şu alanları içersin:
  - ad
  - zorluk
  - odul

Programdan istenenler:

A) Başlangıç durumunu yazdırın.
- Tüm oyuncuların isim, seviye, puan ve aktif bilgilerini gösterin.

B) Görev dağıtımı yapın.
- Sadece aktif oyuncular görev alsın.
- Oyuncunun seviyesi görev zorluğuna eşit veya büyükse görev tamamlansın:
  - Oyuncklayın:
  - tamamlanan_sayi
  - ilk_gorev (yoksa None)

D) Envanter kontrolü yapın.
- Envanterde "anahtar" varsa silin.
- Envanter tamamen boşsa varsayılan bir eşya ekleyin.

E) Ceza/ödül adımı ekleyin.
- Puanı 900'ün altında olan oyunculara +100 destek puanı verin.

F) Güvenli oyuncu araması yapın.
- `get()` kullanarak sistemde olmayan bir oyuncuyu arayın ve sonucu yazdırın.

G) Liderlik tablosu oluşturun.
- Oyuncuları puana göre azalan sıralayın (`sorted`, `lambda`).
- Sıralamayı `while` döngüsü ile ekrana yazdırın.

H) Sistemden bir oyuncu çıkarın.
- En düşük puanlı oyuncuyu `pop()` ile çıkarın.
- Çıkan oyuncuyu ve final oyuncu listesini yazdırın.

Kurallar:
- Dictionary, list, `for`, `while`, `if`, `continue`, `pass`, `get`, `setdefault`, `pop`, `sorted`, `lambda` kullanın.
- Kodunuzu açıklayıcı başlıklarla çıktı verecek şekilde yazın.[8:41 PM]bu bizim sorumuz arkadaslar, zor degil ama baya tekrar edeceginiz yerler var.

```python
print("=== MINI OYUN KAFESI ===")

oyuncular = {
    "Ahmet": {"seviye": 3, "puan": 820, "envanter": ["iksir", "anahtar"], "aktif": True},
    "Zeynep": {"seviye": 7, "puan": 960, "envanter": ["kalkan"], "aktif": True},
    "Mehmet": {"seviye": 2, "puan": 450, "envanter": [], "aktif": False},
}
gorevler = [
    {"ad": "Goblin Avi", "zorluk": 2, "odul": 120},
    {"ad": "Boss Savasi", "zorluk": 6, "odul": 350},
]

print("\n--- A) Baslangic ---")
for isim, b in oyuncular.items():
    print(f"{isim} | seviye={b['seviye']} puan={b['puan']} aktif={b['aktif']}")

print("\n--- B/C/D/E) Gorev + Istatistik + Envanter + Destek ---")
for isim, b in oyuncular.items():
    b.setdefault("istatistik", {"tamamlanan_sayi": 0, "ilk_gorev": None})
    if not b["aktif"]:
        print(f"{isim} pasif -> continue"); continue
    for g in gorevler:
        if b["seviye"] >= g["zorluk"]:
            bonus = 25 if "iksir" in b["envanter"] else 0
            b["puan"] += g["odul"] + bonus
            b["istatistik"]["tamamlanan_sayi"] += 1
            if b["istatistik"]["ilk_gorev"] is None: b["istatistik"]["ilk_gorev"] = g["ad"]
    if "anahtar" in b["envanter"]: b["envanter"].remove("anahtar")
    if not b["envanter"]: b["envanter"].append("baslangic_tas")
    if b["puan"] < 900: b["puan"] += 100

print("\n--- F) get ile arama ---")
print("Bilinmeyen:", oyuncular.get("Bilinmeyen"))

print("\n--- G) Liderlik (while) ---")
sirali = sorted(oyuncular.items(), key=lambda it: it[1]["puan"], reverse=True)
i = 0
while i < len(sirali):
    isim, b = sirali[i]
    print(f"{i+1}. {isim} -> {b['puan']} puan | ist={b['istatistik']}")
    i += 1

print("\n--- H) En dusuk puanliyi cikar (pop) ---")
en_dusuk = min(oyuncular.items(), key=lambda it: it[1]["puan"])[0]
cikan = oyuncular.pop(en_dusuk)
print("Cikan:", en_dusuk, cikan)
print("Kalanlar:", list(oyuncular.keys()))
```
