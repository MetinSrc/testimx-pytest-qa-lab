## 1. Kontrol Yapıları Nedir?

Kontrol yapıları, programın akışını yönetmek için kullanılan yapılardır. Python'da temel kontrol yapıları şunlardır:

- **Koşullu İfadeler:** `if`, `elif`, `else`
- **Döngüler:** `for`, `while`
- **Döngü Kontrol İfadeleri:** `break`, `continue`, `pass`
- **Comprehensions:** List, Dictionary, Set comprehensions
- **Ternary Operator:** Kısa if-else yapısı

**Temel Özellikler:**
- Python'da bloklar girinti (indentation) ile belirlenir
- `:` (iki nokta) ile blok başlatılır
- 4 boşluk veya 1 tab kullanılır (tutarlı olmalı)
- Bloklar iç içe geçebilir (nested)

a) if Statement

Belirli bir koşul sağlandığında kod bloğunu çalıştırır.

**Syntax:**
```python
if koşul:
    # Koşul True ise çalışacak kod
    kod_bloğu
```

**Örnek:**
```python
yas = 18
if yas >= 18:
    print("Reşitsiniz")
```

````python
yas = 25

if yas >= 18:
    print("Reşitsiniz")
    
print("Bitti")
````

### b) if-else Statement

Koşul sağlanmazsa alternatif kod bloğunu çalıştırır.

**Syntax:**
```python
if koşul:
    # Koşul True ise çalışacak kod
    kod_bloğu_1
else:
    # Koşul False ise çalışacak kod
    kod_bloğu_2
```

**Örnek:**
```python
yas = 16
if yas >= 18:
    print("Reşitsiniz")
else:
    print("Reşit değilsiniz")
```

````python
yas = 25

if yas >= 18:
    print("Reşitsiniz")
    print("Hayirli olsun")
else:
    print("Reşit değilsiniz")
    
    
print("Bitti")
````

### c) if-elif-else Statement

Birden fazla koşulu sırayla kontrol eder.

**Syntax:**
```python
if koşul1:
    kod_bloğu_1
elif koşul2:
    kod_bloğu_2
elif koşul3:
    kod_bloğu_3
else:
    kod_bloğu_4
```

**Örnek:**
```python
not_puani = 85

if not_puani >= 90:
    print("Harf Notu: A")
elif not_puani >= 80:
    print("Harf Notu: B")
elif not_puani >= 70:
    print("Harf Notu: C")
elif not_puani >= 60:
    print("Harf Notu: D")
else:
    print("Harf Notu: F")
```

````python
harf="B"

if harf == "C":
    print("Senin harfin C")
elif harf =="D":
    print("Senin harfin D")
elif harf =="E":
        print("Senin harfin E")
else:
    print("Soylemek istemiyorum")
````

**Önemli Notlar:**
- `elif` sayısı sınırsızdır
- İlk `True` olan koşulun bloğu çalışır, diğerleri kontrol edilmez
- `else` opsiyoneldir ama en sonda olmalıdır
- Tüm `elif`'ler aynı seviyede olmalıdır

### d) İç İçe if Yapıları (Nested if)

if blokları içinde başka if blokları kullanılabilir.

**Örnek:**
```python
yas = 20
ehliyet = True

if yas >= 18:
    if ehliyet:
        print("Araba kullanabilirsiniz")
    else:
        print("Ehliyetiniz yok")
else:
    print("Yaşınız yeterli değil")
```

**Alternatif (and operatörü ile):**
```python
yas = 20
ehliyet = True

if yas >= 18 and ehliyet:
    print("Araba kullanabilirsiniz")
elif yas >= 18:
    print("Ehliyetiniz yok")
else:
    print("Yaşınız yeterli değil")
```
 

````python
yas = 17
ehliyet = False

if yas >= 18:
    if ehliyet:
        print("Araba kullanabilirsiniz")
    else:
        print("Ehliyetiniz yok")
else:
    print("Yaşınız yeterli değil")
````

````python
print("B" == "B")
print(8>=5)

print("B" != "B")

print(8>5 and "B" =="B")
print(8<5 or "B" == "b")
````

````python
yas = 25
gozluk= False
ehliyet = False

if yas >= 18 and not gozluk :
    if ehliyet:
        print("Araba kullanabilirsiniz")
    else:
        print("Ehliyetiniz yok")
else:
    print("Yaşınız yeterli değil")
````

### e) Mantıksal Operatörler ile Koşullar

**and (ve):**
```python
if yas >= 18 and ehliyet:
    print("Araba kullanabilirsiniz")
```

**or (veya):**
```python
if not_puani >= 90 or not_puani < 60:
    print("Özel durum")
```

**not (değil):**
```python
if not ehliyet:
    print("Ehliyetiniz yok")
```

**Örnek:**
```python
yas = 25
ehliyet = True
araba_var = False

if yas >= 18 and ehliyet and araba_var:
    print("Araba kullanabilirsiniz")
elif yas >= 18 and ehliyet:
    print("Ehliyetiniz var ama arabanız yok")
elif yas >= 18:
    print("Ehliyetiniz yok")
else:
    print("Yaşınız yeterli değil")
```
