## 1. Sınıf ve Nesne

### Sınıf (Class) Nedir?

Sınıf, nesnelerin **şablonu** veya **planı**dır. Gerçek hayattaki bir kavramı modellemek için kullanılır.

- **Sınıf:** Bir kavramın tanımı (örn: Araba, Kişi, Hayvan)
- **Nesne (Object/Instance):** Sınıftan üretilen somut örnekler (örn: Kırmızı Mercedes, Ahmet, Karabaş)

### Nesne (Object) Nedir?

Nesne, bir sınıftan oluşturulan somut bir örnektir. Her nesne:
- **Öznitelikler (attributes):** Veri tutar (isim, yaş, renk vb.)
- **Metodlar (methods):** Davranışları tanımlar (yürümek, konuşmak vb.)

```python
class Arabalar:
    renk="Kirmizi"
    motor_gucu=110
    marka="Opel"
    model="Mocca"
    yakit_turu="Dizel"
    vites="Manuel"
```

```python
araba1 = Arabalar()
araba2 = Arabalar()

print("Araba 1 ",araba1.renk)
print("Araba 1 ",araba1.yakit_turu)

print("Araba 2 ",araba2.renk)
print("Araba 2 ",araba2.yakit_turu)
```

## Instance Variables vs Class Variables

| Tür | Tanım Yeri | Kullanım | Erişim |
|-----|------------|----------|--------|
| **Instance variable** | `__init__` içinde `self.` ile | Her nesneye özel | `nesne.degisken` |
| **Class variable** | Sınıf gövdesinde | Tüm nesneler için ortak | `Sinif.degisken` veya `nesne.degisken` |

```python
class Arabalar:
    
    def __init__(self,renk,motor_gucu,marka,model,yakit_turu,vites):
        self.renk=renk
        self.motor_gucu=motor_gucu
        self.marka=marka
        self.model=model
        self.yakit_turu = yakit_turu
        self.vites = vites
        
    def call_model(self):
        return f"{self.marka} marka aracin {self.model} modelini tanitiyorum..."
```

```python
araba3 = Arabalar("Mavi",112, "Renoult" ,"XC" , "Benzin" , "Manuel")
araba4 = Arabalar("Yesil",220, "Tesla" ,"XC" , "Elektrikli" , "Auto")

print(araba3.marka)
print(araba3.call_model())

print(araba4.marka)
print(araba4.call_model())
```

```python
class BankaHesabi:
    bakiye = 0
    def __init__(self,hesap_no="Tanimlanmadi",bakiye=0):
        self.hesap_no=hesap_no
        self.bakiye=bakiye
        self.cvc="123"
        
    def para_yatir(self,miktar):
        BankaHesabi.bakiye += miktar
        self.bakiye+=miktar
        print(f"Yatirma islemi tamamlanmistir. Bakiyeniz : {BankaHesabi.bakiye} {self.bakiye}")
        
```

```python
hesap2= BankaHesabi(bakiye=500000, hesap_no="TR3454465765786")
hesap= BankaHesabi(bakiye=300000, hesap_no="TR3454465765786")

print(hesap.cvc)
print(hesap.bakiye)
print("Hesap 2 :" , hesap2.bakiye)
print(hesap.hesap_no)

hesap.para_yatir(20000)
hesap2.para_yatir(75000)

print(hesap.bakiye)
print(hesap2.bakiye)
```
