# Python Veri Yapıları - Detaylı Ders Notu

Bu notta Python'daki temel veri yapıları olan `list`, `tuple`, `set` ve `dictionary` konularını detaylı ve örnekli şekilde inceleyeceğiz.

## 1. Veri Yapıları Neden Önemli?

- Veriyi düzenli saklamayı sağlar.
- Problemleri daha hızlı ve okunabilir çözmeye yardım eder.
- Fonksiyonlarla birlikte kullanıldığında kodu güçlü ve esnek hale getirir.

## 2. List (Liste)

Liste, sıralı ve değiştirilebilir (mutable) bir veri yapısıdır.

### a) Liste Oluşturma

```python
sayilar = [10, 20, 30]
karisik = [1, "Python", 3.14, True]
bos_liste = []
```

### b) Liste Özellikleri

- Sıralıdır (index vardır).
- Değiştirilebilir.
- Aynı elemandan birden fazla içerebilir.
- Farklı veri tiplerini birlikte tutabilir.

### c) Temel Liste Metotları

```python
meyveler = ["elma", "armut"]
meyveler.append("muz")            # Sona ekler
meyveler.insert(1, "çilek")       # Belirli indexe ekler
meyveler.extend(["kiraz", "kivi"])# Listeyi genişletir
meyveler.remove("armut")          # Değere göre siler
silinen = meyveler.pop()          # Son elemanı siler ve döndürür
```

### d) Liste Üzerinde Döngü

```python
notlar = [70, 85, 90, 60]
toplam = 0
for notu in notlar:
    toplam += notu
print("Ortalama:", toplam / len(notlar))
```

### e) List Comprehension

```python
kareler = [x * x for x in range(1, 6)]
ciftler = [x for x in range(20) if x % 2 == 0]
```

## 3. Tuple (Demet)

Tuple, sıralı ama değiştirilemeyen (immutable) bir veri yapısıdır.

### a) Tuple Oluşturma

```python
koordinat = (10, 20)
renkler = ("kırmızı", "yeşil", "mavi")
tek_eleman = (5,)  # Virgül önemli
```

### b) Tuple Özellikleri

- Sıralıdır.
- Değiştirilemez.
- Aynı elemandan birden fazla olabilir.
- Listelere göre daha güvenli ve bazı durumlarda daha hızlıdır.

### c) Tuple Kullanım Alanları

- Sabit veri saklama (örn. gün adları)
- Fonksiyondan birden fazla değer döndürme

```python
def min_max(liste):
    return (min(liste), max(liste))

sonuc = min_max([3, 9, 1, 6])
print(sonuc)      # (1, 9)
print(sonuc[0])   # 1
```

### d) Tuple Metotları ve Fonksiyon Örnekleri

Tuple, listeye göre daha az metoda sahiptir. Kendi metotları:
- `count()`
- `index()`

```python
sayilar = (10, 20, 30, 20, 40, 20)

print(sayilar.count(20))  # 3 (20 değeri kaç kez var?)
print(sayilar.index(30))  # 2 (30 değeri ilk hangi indexte?)
```

Tuple ile sık kullanılan built-in fonksiyonlar:

```python
notlar = (70, 85, 90, 60)

print(len(notlar))     # 4
print(min(notlar))     # 60
print(max(notlar))     # 90
print(sum(notlar))     # 305
print(sum(notlar) / len(notlar))  # 76.25
```

`sorted()` tuple'ı sıralayıp liste döndürür:

```python
degerler = (5, 1, 8, 3)
sirali_liste = sorted(degerler)
print(sirali_liste)  # [1, 3, 5, 8]
print(type(sirali_liste))  # <class 'list'>
```

Tuple içinde üyelik kontrolü:

```python
renkler = ("kirmizi", "mavi", "yesil")
print("mavi" in renkler)      # True
print("sari" not in renkler)  # True
```

## 4. Set (Küme)

Set, sırasız ve benzersiz elemanlardan oluşan veri yapısıdır.

### a) Set Oluşturma

```python
sayilar = {1, 2, 3, 4}
bos_set = set()  # {} dictionary olur, dikkat
```

### b) Set Özellikleri

- Sırasızdır (index yoktur).
- Aynı elemandan birden fazla tutmaz.
- Değiştirilebilir.

```python
tekrarli = [1, 2, 2, 3, 3, 3, 4]
benzersiz = set(tekrarli)
print(benzersiz)  # {1, 2, 3, 4}
```

### c) Set Metotları

```python
harfler = {"a", "b", "c"}
harfler.add("d")
harfler.remove("b")   # Yoksa hata verir
harfler.discard("x")  # Yoksa hata vermez
```

### d) Küme İşlemleri

```python
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print(a | b)  # Birleşim
print(a & b)  # Kesişim
print(a - b)  # Fark
print(a ^ b)  # Simetrik fark
```

## 5. Dictionary (Sözlük)

Dictionary, anahtar-değer (key-value) mantığıyla çalışan veri yapısıdır.

### a) Dictionary Oluşturma

```python
ogrenci = {
    "isim": "Ayşe",
    "yas": 16,
    "sinif": "10-A"
}
```

### b) Dictionary Özellikleri

- Anahtarlar benzersizdir.
- Değerler tekrar edebilir.
- Python 3.7+ sürümlerinde eklenme sırasını korur.
- Değiştirilebilir.

### c) Temel Dictionary İşlemleri

```python
print(ogrenci["isim"])         # Ayşe
print(ogrenci.get("numara"))   # None

ogrenci["yas"] = 17            # Güncelleme
ogrenci["okul"] = "Anadolu"    # Yeni anahtar

del ogrenci["sinif"]           # Silme
```

### d) items(), keys(), values()

```python
urun = {"ad": "Kalem", "fiyat": 15, "stok": 120}

for anahtar in urun.keys():
    print(anahtar)

for deger in urun.values():
    print(deger)

for anahtar, deger in urun.items():
    print(f"{anahtar}: {deger}")
```

### e) Dictionary Comprehension

```python
kare_sozluk = {x: x * x for x in range(1, 6)}
```
