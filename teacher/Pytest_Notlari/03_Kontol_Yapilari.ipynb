## 1. Kontrol Yapıları Nedir?

Kontrol yapıları, programın akışını yönetmek için kullanılan yapılardır. Python'da temel kontrol yapıları şunlardır:

- **Koşullu İfadeler:** `if`, `elif`, `else`
- **Döngüler:** `for`, `while`
- **Döngü Kontrol İfadeleri:** `break`, `continue`, `pass`
- **Comprehensions:** List, Dictionary, Set comprehensions
- **Ternary Operator:** Kısa if-else yapısı

**Temel Özellikler:**
- Python'da bloklar girinti (indentation) ile belirlenir
- `:` (iki nokta) ile blok başlatılır
- 4 boşluk veya 1 tab kullanılır (tutarlı olmalı)
- Bloklar iç içe geçebilir (nested)

a) if Statement

Belirli bir koşul sağlandığında kod bloğunu çalıştırır.

**Syntax:**
```python
if koşul:
    # Koşul True ise çalışacak kod
    kod_bloğu
```

**Örnek:**
```python
yas = 18
if yas >= 18:
    print("Reşitsiniz")
```

````python
yas = 25

if yas >= 18:
    print("Reşitsiniz")
    
print("Bitti")
````

### b) if-else Statement

Koşul sağlanmazsa alternatif kod bloğunu çalıştırır.

**Syntax:**
```python
if koşul:
    # Koşul True ise çalışacak kod
    kod_bloğu_1
else:
    # Koşul False ise çalışacak kod
    kod_bloğu_2
```

**Örnek:**
```python
yas = 16
if yas >= 18:
    print("Reşitsiniz")
else:
    print("Reşit değilsiniz")
```

````python
yas = 25

if yas >= 18:
    print("Reşitsiniz")
    print("Hayirli olsun")
else:
    print("Reşit değilsiniz")
    
    
print("Bitti")
````

### c) if-elif-else Statement

Birden fazla koşulu sırayla kontrol eder.

**Syntax:**
```python
if koşul1:
    kod_bloğu_1
elif koşul2:
    kod_bloğu_2
elif koşul3:
    kod_bloğu_3
else:
    kod_bloğu_4
```

**Örnek:**
```python
not_puani = 85

if not_puani >= 90:
    print("Harf Notu: A")
elif not_puani >= 80:
    print("Harf Notu: B")
elif not_puani >= 70:
    print("Harf Notu: C")
elif not_puani >= 60:
    print("Harf Notu: D")
else:
    print("Harf Notu: F")
```

````python
harf="B"

if harf == "C":
    print("Senin harfin C")
elif harf =="D":
    print("Senin harfin D")
elif harf =="E":
        print("Senin harfin E")
else:
    print("Soylemek istemiyorum")
````

**Önemli Notlar:**
- `elif` sayısı sınırsızdır
- İlk `True` olan koşulun bloğu çalışır, diğerleri kontrol edilmez
- `else` opsiyoneldir ama en sonda olmalıdır
- Tüm `elif`'ler aynı seviyede olmalıdır

### d) İç İçe if Yapıları (Nested if)

if blokları içinde başka if blokları kullanılabilir.

**Örnek:**
```python
yas = 20
ehliyet = True

if yas >= 18:
    if ehliyet:
        print("Araba kullanabilirsiniz")
    else:
        print("Ehliyetiniz yok")
else:
    print("Yaşınız yeterli değil")
```

**Alternatif (and operatörü ile):**
```python
yas = 20
ehliyet = True

if yas >= 18 and ehliyet:
    print("Araba kullanabilirsiniz")
elif yas >= 18:
    print("Ehliyetiniz yok")
else:
    print("Yaşınız yeterli değil")
```
 

````python
yas = 17
ehliyet = False

if yas >= 18:
    if ehliyet:
        print("Araba kullanabilirsiniz")
    else:
        print("Ehliyetiniz yok")
else:
    print("Yaşınız yeterli değil")
````

````python
print("B" == "B")
print(8>=5)

print("B" != "B")

print(8>5 and "B" =="B")
print(8<5 or "B" == "b")
````

````python
yas = 25
gozluk= False
ehliyet = False

if yas >= 18 and not gozluk :
    if ehliyet:
        print("Araba kullanabilirsiniz")
    else:
        print("Ehliyetiniz yok")
else:
    print("Yaşınız yeterli değil")
````

## 4. For Döngüsü

Belirli bir sayıda veya bir koleksiyonun elemanları üzerinde tekrar eden işlemler için kullanılır.

### a) Temel For Döngüsü

**Syntax:**
```python
for eleman in iterable:
    kod_bloğu
```

**Örnekler:**
```python
# String üzerinde döngü
for harf in "Python":
    print(harf)
# Çıktı: P, y, t, h, o, n

# Liste üzerinde döngü
meyveler = ["elma", "armut", "muz"]
for meyve in meyveler:
    print(meyve)
# Çıktı: elma, armut, muz

# Tuple üzerinde döngü
sayilar = (1, 2, 3, 4, 5)
for sayi in sayilar:
    print(sayi * 2)
# Çıktı: 2, 4, 6, 8, 10

````python
text="Python"

for i in text:
    if i=="t" :
        break
    
    print(i)
````

````python
text1="Python"
text2="Java"

for i in text1:
    for j in text2:
        if j=="v" :
            break
    
        print(j)
    
    print(i + "-----" +j)
````

````python
meyveler = ["elma", "armut", "muz"]
for meyve in meyveler:
    for i in meyve:
        print(i)
````

````python
sesli="aEioU".lower()

harf =input("Lutfen bir harf giriniz").lower()


for i in sesli:
    if harf==i :
        print("Sesli harfi sectiniz")
    else :
        print("Sessiz harf girdiniz")
        
    break
    
        
````

range() Fonksiyonu ile For Döngüsü

`range()` belirli bir aralıktaki sayıları üretir.

**Syntax:**
```python
range(stop)              # 0'dan stop'a kadar (stop hariç)
range(start, stop)       # start'tan stop'a kadar (stop hariç)
range(start, stop, step) # start'tan stop'a kadar step adımla
```

````python
for i in range(3,22,3):
    print(i)
````

 c) enumerate() ile For Döngüsü

Hem index hem de değeri almak için kullanılır.

**Syntax:**
```python
for index, değer in enumerate(iterable):
    kod_bloğu
```

**Örnek:**
```python
meyveler = ["elma", "armut", "muz"]

# enumerate olmadan
for i in range(len(meyveler)):
    print(f"{i}: {meyveler[i]}")

# enumerate ile (daha Pythonic)
for i, meyve in enumerate(meyveler):
    print(f"{i}: {meyve}")
# Çıktı:
# 0: elma
# 1: armut
# 2: muz

# Başlangıç index'i belirleme
for i, meyve in enumerate(meyveler, start=1):
    print(f"{i}: {meyve}")
# Çıktı:
# 1: elma
# 2: armut
# 3: muz
```

````python
text="Python"

for i in enumerate(text,3):
    print(i)
````

````python
meyveler = ["armut", "elma",  "muz"]

for sayi, meyve in enumerate(meyveler, start=2): 
    print(f"{sayi}: {meyve}")
````

### d) zip() ile For Döngüsü

Birden fazla iterable'ı aynı anda döngüye almak için kullanılır.

**Syntax:**
```python
for eleman1, eleman2 in zip(iterable1, iterable2):
    kod_bloğu
```

**Örnek:**
```python
isimler = ["Ali", "Veli", "Ayşe"]
yaslar = [25, 30, 28]

for isim, yas in zip(isimler, yaslar):
    print(f"{isim} {yas} yaşında")
# Çıktı:
# Ali 25 yaşında
# Veli 30 yaşında
# Ayşe 28 yaşında
```

**Önemli Not:**
- `zip()` en kısa iterable bitince durur
- Farklı uzunluklarda iterable'lar için `itertools.zip_longest()` kullanılabilir

````python
ogrenciler=["Ali","Ayse", "Handan"]
no=[216,217,218]

for i , j in zip(ogrenciler,no):
    print(f"{i}'nin numarasi {j}")
````

````python
ogrenciler = ["Ali", "Ayşe", "Mehmet", "Zeynep"]
notlar = [85, 92, 78]
dersler = ["Matematik", "Fizik", "Kimya"]

for i in ogrenciler:
    print("Ogrencinin adi: ", i)
    for j,k in zip(dersler,notlar):
        print("Ders adi: ", j)
        print("Not: ", k)
        print("*******SONUC********")
        print(f"{i}nin {j} ders notu {k}")
        print("***********************\n")
  
````

## 5. While Döngüsü

Koşul sağlandığı sürece tekrar eden işlemler için kullanılır.

### a) Temel While Döngüsü

**Syntax:**
```python
while koşul:
    kod_bloğu
```

**Örnek:**
```python
sayac = 0
while sayac < 5:
    print(sayac)
    sayac += 1
# Çıktı: 0, 1, 2, 3, 4
```

### b) Sonsuz Döngü ve Kontrolü

**Sonsuz Döngü:**
```python
while True:
    # Sonsuz döngü
    kullanici_girdisi = input("Çıkmak için 'q' yazın: ")
    if kullanici_girdisi == 'q':
        break
```

**Örnek: Sayı Tahmin Oyunu**
```python
hedef = 42
tahmin = None

while tahmin != hedef:
    tahmin = int(input("Tahmininiz: "))
    if tahmin < hedef:
        print("Daha büyük!")
    elif tahmin > hedef:
        print("Daha küçük!")
    else:
        print("Tebrikler! Doğru tahmin!")
```

````python
hedef = 42
tahmin = 42

while tahmin != hedef:
    print(tahmin)
    tahmin = int(input("Tahmininiz: "))
    if tahmin < hedef:
        print("Daha büyük!")
    elif tahmin > hedef:
        print("Daha küçük!")
    else:
        print("Tebrikler! Doğru tahmin!")
  
  


 
````
