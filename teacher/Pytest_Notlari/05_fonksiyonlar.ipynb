# Fonksiyonlar (Functions) - Detaylı Ders Notu

Fonksiyonlar, belirli bir işi yapan ve tekrar tekrar kullanılabilen kod bloklarıdır.

## 1. Fonksiyon Neden Kullanılır?

- **Tekrarı azaltır:** Aynı kodu her yerde yeniden yazmazsınız.
- **Okunabilirliği artırır:** Kodunuz anlamlı parçalara ayrılır.
- **Bakımı kolaylaştırır:** Hata düzeltme tek bir yerde yapılır.
- **Test etmeyi kolaylaştırır:** Her fonksiyon ayrı ayrı test edilebilir.

## 2. Temel Fonksiyon Tanımı

**Syntax:**
```python
def fonksiyon_adi(parametre1, parametre2):
    # Kod bloğu
    return sonuc
```

**Örnek:**
```python
def selamla(isim):
    return f"Merhaba, {isim}!"

mesaj = selamla("Ayşe")
print(mesaj)  # Merhaba, Ayşe!
```

## 3. Parametre ve Argüman Kavramı

- **Parametre:** Fonksiyon tanımında yer alır (`def topla(a, b):`)
- **Argüman:** Fonksiyon çağrılırken gönderilen gerçek değer (`topla(3, 5)`)

**Örnek:**
```python
def topla(a, b):
    return a + b

print(topla(10, 20))  # 30
```

## 4. Fonksiyon Türleri

### a) Parametresiz, Değer Döndürmeyen Fonksiyon
```python
def kurallari_yazdir():
    print("1. Derse zamanında gel")
    print("2. Ödevleri tamamla")

kurallari_yazdir()
```

### b) Parametreli, Değer Döndürmeyen Fonksiyon
```python
def ogrenci_bilgisi(isim, yas):
    print(f"Öğrenci: {isim}, Yaş: {yas}")

ogrenci_bilgisi("Mehmet", 17)
```

### c) Parametresiz, Değer Döndüren Fonksiyon
```python
def okul_adi():
    return "Atatürk Anadolu Lisesi"

print(okul_adi())
```

### d) Parametreli, Değer Döndüren Fonksiyon
```python
def not_ortalamasi(vize, final):
    return vize * 0.4 + final * 0.6

ortalama = not_ortalamasi(70, 80)
print(ortalama)  # 76.0
```

## 5. return vs print

- `print()` sadece ekrana yazar.
- `return` değeri fonksiyon dışına gönderir.

**Örnek karşılaştırma:**
```python
def kare_print(x):
    print(x * x)

def kare_return(x):
    return x * x

sonuc = kare_return(5)
print(sonuc + 10)  # 35
```

## 8. *args ve **kwargs

### a) *args (Sınırsız pozisyonel argüman)
```python
def ortalama_hesapla(*notlar):
    if len(notlar) == 0:
        return 0
    return sum(notlar) / len(notlar)

print(ortalama_hesapla(70, 80, 90))  # 80.0
```

### b) **kwargs (Sınırsız anahtar-değer argümanı)
```python
def ogrenci_karti(**bilgiler):
    for anahtar, deger in bilgiler.items():
        print(f"{anahtar}: {deger}")

ogrenci_karti(isim="Ece", sinif="10-A", numara=123)
```
