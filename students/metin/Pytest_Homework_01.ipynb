### Soru 24: KDV Hesaplama
KDV hesaplama programı yazınız:
- Temel gıda: %1
- Diğer gıda: %10
- Kitap: %8
- Temizlik: %18
- Elektronik: %20
- Diğer: %18
Kullanıcı ürün tipini seçsin ve fiyata göre KDV'li fiyatı hesaplayın.

```python
# Kodunuzu buraya yazınız
urun_fiyatı= input("Ürün fiyatını giriniz: ")
urun_turu= input("Ürün türünü giriniz (Temel gıda,Diger gıda, Kitap, Temizlik, Elektronik, Diger): ")

if urun_fiyatı.isdigit():
    urun_fiyatı=float(urun_fiyatı)

    if urun_turu=="Temel gıda":
        urun_fiyatı= urun_fiyatı*1.01
        print("KDV Eklenmiş  Yeni fiyat:", urun_fiyatı)

    elif urun_turu=="Diger gıda":
        urun_fiyatı= urun_fiyatı *1.1
        print("KDV Eklenmiş  Yeni fiyat:", urun_fiyatı)

    elif urun_fiyatı=="Kitap":
        urun_fiyatı= urun_fiyatı*1.08
        print("KDV Eklenmiş  Yeni fiyat:", urun_fiyatı)

    elif urun_turu=="Temizlik":
        urun_fiyatı= urun_fiyatı*1.18
        print("KDV Eklenmiş  Yeni fiyat:", urun_fiyatı)

    elif urun_turu=="Elektronik":
        urun_fiyatı= urun_fiyatı*1.20
        print("KDV Eklenmiş  Yeni fiyat:", urun_fiyatı)

    elif urun_turu=="Diger":
        urun_fiyatı= urun_fiyatı*1.18
        print("KDV Eklenmiş  Yeni fiyat:", urun_fiyatı)
    else :
        print("Lütfen geçerli bir ürün türü giriniz.")
```

```python
# Yapay Zekanın Çözümü
kdv_oranlari = {
    "temel gıda": 0.01,
    "diger gıda": 0.10,
    "kitap": 0.08,
    "temizlik": 0.18,
    "elektronik": 0.20,
    "diger": 0.18
}

print("Mevcut ürün türleri:")
urun_listesi = list(kdv_oranlari.keys())

for i, urun in enumerate(urun_listesi, 1):
    print(f"{i}) {urun}")

urun_fiyati = float(input("\nÜrün fiyatı: "))
secim = int(input("Ürün numarası seçiniz: "))

if 1 <= secim <= len(urun_listesi):
    urun_turu = urun_listesi[secim - 1]
    yeni_fiyat = urun_fiyati * (1 + kdv_oranlari[urun_turu])
    print(f"Yeni fiyat: {yeni_fiyat:.2f} TL")
else:
    print("Geçersiz seçim.")
```

### Soru 27: Maaş Hesaplama
Bir iş yerinde çalışanın maaşını hesaplayan program yazınız:
- Saatlik ücret: 100₺
- Haftalık 45 saate kadar: normal ücret
- 45-55 saat arası: 1.5 katı
- 55 saat üzeri: 2 katı
- Haftada 70 saatten fazla çalıştırılamaz

```python
# Kodunuzu buraya yazınız
saatlik_ucret = 100
calisma_suresi = int(input("Çalışma süresini tam sayi cinsinden giriniz: "))

normal_ucret = saatlik_ucret * calisma_suresi
tam_mesai_ucreti = saatlik_ucret * 45

if 45 < calisma_suresi <= 55:
    ucret = tam_mesai_ucreti + (calisma_suresi - 45) * saatlik_ucret * 1.5
    print("1. Seviye fazla mesai uygulandı:", ucret)

elif 55 < calisma_suresi <= 70:
    ucret = tam_mesai_ucreti + (10 * saatlik_ucret * 1.5) + (calisma_suresi - 55) * saatlik_ucret * 2
    print("1. ve 2. Seviye fazla mesai uygulandı:", ucret)

elif calisma_suresi > 70:
    print("70 Saatten fazla çalışma süresi olamaz.")

elif calisma_suresi < 0:
    print("Geçersiz süre.")

else:
    print(f"Normal ücret: {normal_ucret} TL")
```

```python
#yapay zekanın çözümü
RATE = 100
MAX_HOURS = 70

hours = float(input("Haftalık çalışma saati: "))

if hours < 0:
    print("Hata: Negatif saat olamaz.")
elif hours > MAX_HOURS:
    print("Hata: Haftada 70 saatten fazla çalışılamaz.")
else:
    normal = min(hours, 45)
    ot1 = max(0, min(hours, 55) - 45)     # 45–55 arası (max 10 saat)
    ot2 = max(0, hours - 55)              # 55 üstü

    pay = (normal * RATE) + (ot1 * RATE * 1.5) + (ot2 * RATE * 2)
    print(f"Toplam maaş: {pay:.2f} TL")
```

### Soru 28: Gelişmiş Not Hesaplama
Not hesaplama sistemi programı yazınız:
- Vize: %30
- Ödev: %20
- Final: %50
- Geçme notu: 60
- Final notu 45'in altındaysa direkt kalır
- Devamsızlık 10 günü geçerse devamsızlıktan kalır

```python
# Kodunuzu buraya yazınız
vize_notu = int(input("Vize notunuzu giriniz: "))
if vize_notu < 0 or vize_notu > 100:
    print("Lütfen 0 ile 100 arasında bir vize notu giriniz.")
ödev_notu = int(input("Ödev notunuzu giriniz: "))
final_notu = int(input("Final notunuzu giriniz: "))
devamsizlik = int(input("Devamsızlık gün sayısını giriniz: "))

ortalama = vize_notu * 0.3 + ödev_notu * 0.2 + final_notu * 0.5

if final_notu < 45:
    print("Sınıfta kaldınız. Final notunuz 45'in altında.")
elif devamsizlik > 10:
    print("Sınıfta kaldınız. Devamsızlık gün sayısı 10'un üzerinde.")
elif ortalama >= 60:
    print("Tebrikler, sınıfı geçtiniz. Ortalama notunuz:", ortalama)
else:
    print("Sınıfta kaldınız. Ortalama notunuz:", ortalama)
```

```python
def sayi_al(mesaj, min_deger, max_deger):
    while True:
        try:
            x = float(input(mesaj))
            if min_deger <= x <= max_deger:
                return x
            print(f"Hata: {min_deger}-{max_deger} arasında olmalı.")
        except ValueError:
            print("Hata: Sayı giriniz.")

# Girişler
vize = sayi_al("Vize (0-100): ", 0, 100)
odev = sayi_al("Ödev (0-100): ", 0, 100)
final = sayi_al("Final (0-100): ", 0, 100)
devamsizlik = sayi_al("Devamsızlık (0-14): ", 0, 14)

# Kurallar
if devamsizlik > 10:
    print("KALDI → Devamsızlık sınırı aşıldı.")

elif final < 45:
    print("KALDI → Final barajı geçilemedi.")

else:
    ort = 0.3*vize + 0.2*odev + 0.5*final
    durum = "GEÇTİ" if ort >= 60 else "KALDI"
    print(f"{durum} → Ortalama: {ort:.2f}")
```

```python
def read_score(label: str) -> float:
    while True:
        try:
            x = float(input(f"{label} (0-100): ").replace(",", "."))
            if 0 <= x <= 100:
                return x
            print("Hata: 0 ile 100 arasında olmalı.")
        except ValueError:
            print("Hata: Sayı giriniz.")

def read_absence() -> int:
    while True:
        try:
            d = int(input("Devamsızlık günü: "))
            if d >= 0:
                return d
            print("Hata: Negatif olamaz.")
        except ValueError:
            print("Hata: Tam sayı giriniz.")

def calculate_average(vize: float, odev: float, final: float) -> float:
    return 0.30*vize + 0.20*odev + 0.50*final

vize = read_score("Vize")
odev = read_score("Ödev")
final = read_score("Final")
devamsizlik = read_absence()

if devamsizlik > 10:
    print("KALDI: Devamsızlıktan kaldı (10 günü geçti).")

elif final < 45:
    print("KALDI: Final barajı (45) geçilemedi.")

else:
    ortalama = calculate_average(vize, odev, final)
    durum = "GEÇTİ" if ortalama >= 60 else "KALDI"
    print(f"{durum}: Ortalama = {ortalama:.2f}")
```

```python
# Kodunuzu buraya yazınız
print("Kahve Menüsü:")
print("1 - Türk kahvesi (25 TL)")
print("2 - Espresso (30 TL)")
print("3 - Americano (35 TL)")
print("4 - Latte (40 TL)")

secim = int(input("Seçiminizi yapınız (1-4): "))

if secim == 1:
    fiyat = 25
elif secim == 2:
    fiyat = 30
elif secim == 3:
    fiyat = 35
elif secim == 4:
    fiyat = 40
else:
    print("Geçersiz seçim.")
    fiyat = 0

if fiyat != 0:
    adet5  = int(input("Kaç tane 5 TL?: "))
    adet10 = int(input("Kaç tane 10 TL?: "))
    adet20 = int(input("Kaç tane 20 TL?: "))
    adet50 = int(input("Kaç tane 50 TL?: "))

    toplam_para = adet5*5 + adet10*10 + adet20*20 + adet50*50

    if toplam_para >= fiyat:
        print("Kahveniz hazırlanıyor ☕")
        print("Toplam para:", toplam_para, "TL")
        print("Para üstü:", toplam_para - fiyat, "TL")
    else:
        print("Yetersiz para.")
        print("Toplam para:", toplam_para, "TL")
        print("Eksik miktar:", fiyat - toplam_para, "TL")
```

### Soru 30: Vücut Kitle İndeksi
Vücut Kitle İndeksi hesaplama programı yazınız: Kilo/(boy*boy) - boy metre cinsinden.
Sonuca göre:
- 18.5 altı: Zayıf
- 18.5-24.9: Normal
- 25-29.9: Fazla kilolu
- 30-34.9: 1. derece obez
- 35-39.9: 2. derece obez
- 40 üstü: 3. derece obez

```python
# Kodunuzu buraya yazınız
kilo = float(input("Kilonuzu kg cinsinden giriniz: "))
boy = float(input("Boyunuzu metre cinsinden giriniz: "))

vucud_kitle_indeksi = kilo / (boy ** 2)
print(f"Vücut Kitle İndeksiniz: {vucud_kitle_indeksi:.2f}")

if vucud_kitle_indeksi < 18.5:
    print("Zayıf")
elif 18.5 <= vucud_kitle_indeksi < 25:
    print("Normal")
elif 25 <= vucud_kitle_indeksi < 30:
    print("Fazla kilolu")
elif 30 <= vucud_kitle_indeksi < 35:
    print("Obez (Sınıf I)")
elif 35 <= vucud_kitle_indeksi < 40:
    print("Obez (Sınıf II)")


elif vucud_kitle_indeksi >= 40:
    print("Obez (Sınıf III)")
```
